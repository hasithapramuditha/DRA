```python
import pandas as pd

# Load data
survey = pd.read_csv('student_survey.csv')

# Display first 5 rows
print("First 5 rows:")
print(survey.head())

# Check data info
print("\nData info:")
print(survey.info())

# Summarize numerical columns
print("\nSummary statistics:")
print(survey.describe())
```

```python
# Check for missing values
print("Missing values:")
print(survey.isnull().sum())

# Handle missing values
survey['Age'] = survey['Age'].fillna(survey['Age'].median())
survey['Test_Score'] = survey['Test_Score'].fillna(survey['Test_Score'].mean())

# Check for duplicates
print("\nNumber of duplicates:", survey.duplicated().sum())
survey = survey.drop_duplicates()

# Fix Age outlier
survey.loc[survey['Age'] > 18, 'Age'] = survey['Age'].median()

# Verify cleaned data
print("\nMissing values after cleaning:")
print(survey.isnull().sum())
print("\nSummary statistics after cleaning:")
print(survey.describe())
```

```python
# Average calculations
print("Average Test Score:", survey['Test_Score'].mean())
print("Average Study Hours:", survey['Hours_Study_Per_Day'].mean())

# Count favorite subjects
print("\nFavorite Subjects Count:")
print(survey['Favorite_Subject'].value_counts())

# Count transportation methods
print("\nTransportation Methods Count:")
print(survey['Transportation'].value_counts())

# Max and min attendance
print("\nMax Attendance Days:", survey['Attendance_Days'].max())
print("Min Attendance Days:", survey['Attendance_Days'].min())
```

```python
import matplotlib.pyplot as plt

# Bar chart for Favorite Subjects
subject_counts = survey['Favorite_Subject'].value_counts()
plt.figure(figsize=(8, 5))
plt.bar(subject_counts.index, subject_counts.values, color='skyblue')
plt.title('Most Popular Subjects Among Students', fontsize=14)
plt.xlabel('Subject', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)
plt.show()
```

```python
# Line chart for Test Scores vs Attendance Days
attendance_scores = survey.groupby('Attendance_Days')['Test_Score'].mean()
plt.figure(figsize=(8, 5))
plt.plot(attendance_scores.index, attendance_scores.values, color='purple')
plt.title('Average Test Score by Attendance Days', fontsize=14)
plt.xlabel('Attendance Days', fontsize=12)
plt.ylabel('Average Test Score', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Histogram for Age distribution
plt.figure(figsize=(8, 5))
plt.hist(survey['Age'], bins=[14.5, 15.5, 16.5, 17.5, 18.5], color='lightgreen',edgecolor='black', linewidth=1.5)
plt.title('Distribution of Student Ages', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Number of Students', fontsize=12)
plt.show()
```

```python
# Scatter plot for Study Hours vs Test Scores
plt.figure(figsize=(8, 5))
plt.scatter(survey['Hours_Study_Per_Day'], survey['Test_Score'], color='coral', alpha=0.7)
plt.title('Study Hours vs Test Scores', fontsize=14)
plt.xlabel('Hours of Study Per Day', fontsize=12)
plt.ylabel('Test Score', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
